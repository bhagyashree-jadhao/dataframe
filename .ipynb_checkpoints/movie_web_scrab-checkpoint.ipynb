```python
#final_movie = pd.DataFrame(['Movie_Name','Released_Year','Director','Movie_Rating','User_voting','Movie_details_Link'])
final_movie = pd.DataFrame(['Movie_Name','Released_Year'])
for i in range(0, len(movie_list)):
    d={}
    #temp_df = pd.DataFrame(columns=['Movie_Name','Released_Year','Director','Movie_Rating','User_voting','Movie_details_Link'])
    temp_df = pd.DataFrame(columns=['Movie_Name','Released_Year'])
    name=movie_list[i].get_text()
    m_string=(' '.join(name.split()).replace('.',' '))
    movie_name=m_string[len(str(i))+1:-7:]
    #print(m_string)
    year=re.search('\d{4}',m_string).group()
    #print(year)
    d['Movie_Name'] = movie_name
    d['Released_Year'] = year
   
    df = temp_df.append(d,ignore_index=True)
    final_movie = final_movie.append(df)
final_movie[]

                           

   
```
