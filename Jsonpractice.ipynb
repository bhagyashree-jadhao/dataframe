```python
import pandas
import numpy
import json
import os
```

```python
def check_json_file(path):
    json_files=[]
    for files in os.listdir(path):
        if os.path.isfile(files):
            if files.endswith(".json"):
                json_files.append(os.path.join(path,files))
    return json_files
```
